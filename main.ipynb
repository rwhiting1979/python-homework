```python
import csv

# Initialize empty lists to hold the menu data and sales data
menu = []
sales = []

# Read in the menu data
with open("menu_data.csv", "r") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

    # Skip the header row
    next(csvreader)

    # Loop through all the rows in the CSV and append them to the menu list
    for row in csvreader:
        menu.append(row)

# Read in the sales data
with open("sales_data.csv", "r") as csvfile:
    csvreader = csv.reader(csvfile, delimiter=",")

    # Skip the header row
    next(csvreader)

    # Loop through all the rows in the CSV and append them to the sales list
    for row in csvreader:
        sales.append(row)

# Initialize an empty report dictionary to hold the per-product results
report = {}

# Loop through every row in the sales list
for row in sales:
    # Get the quantity, sales item, and price from the sales data
    quantity = int(row[3])
    sales_item = row[4]
    price = 0

    # Check if the sales item is already in the report
    if sales_item not in report:
        # If not, add it to the report with default values
        report[sales_item] = {
            "01-count": 0,
            "02-revenue": 0,
            "03-cogs": 0,
            "04-profit": 0,
        }

    # Loop through every row in the menu list
    for menu_row in menu:
        # Get the item, price, and cost from the menu data
        item = menu_row[0]
        price = float(menu_row[3])
        cost = float(menu_row[4])

        # Check if the sales item and menu item match
        if sales_item == item:
            # If they match, calculate the profit for this item
            profit = (price - cost) * quantity
            # Add the values to the report
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += profit
            break
        else:
            # If they don't match, print a message
            print(f"{sales_item} does not equal {item}! NO MATCH!")

# Write the contents of the report dictionary to a text file
with open("report.txt", "w") as f:
    for k, v in report.items():
        f.write(f"{k}: {v}\n")
```
